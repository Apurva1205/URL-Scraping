```python
import requests
from bs4 import BeautifulSoup

for page_num in range(1, 21): # change the range to scrape more or fewer pages
    url = f"https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&page={page_num}&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_{page_num}"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    
    product_list = soup.find_all("div", class_="s-result-item")
    
    for product in product_list:
        # Extract the product information
        try:
            product_url = "https://www.amazon.in" + product.find("a", class_="a-link-normal")["href"]
        except:
            product_url = ""
        
        try:
            product_name = product.find("h2", class_="a-size-mini").text.strip()
        except:
            product_name = ""
            
        try:
            product_price = product.find("span", class_="a-price-whole").text.strip()
        except:
            product_price = ""
        
        try:
            product_rating = product.find("span", class_="a-icon-alt").text.strip().split()[0]
        except:
            product_rating = ""
            
        try:
            num_reviews = product.find("span", class_="a-size-base").text.strip().split()[0]
        except:
            num_reviews = ""
        
        # Print the product information
        print("Product URL:", product_url)
        print("Product Name:", product_name)
        print("Product Price:", product_price)
        print("Product Rating:", product_rating)
        print("Number of Reviews:", num_reviews)
        print()
```

```python
import requests
from bs4 import BeautifulSoup
import csv

def get_product_info(product_url):
    response = requests.get(product_url)
    soup = BeautifulSoup(response.content, "html.parser")

    try:
        description = soup.find("ul", class_="a-unordered-list").text.strip()
    except:
        description = ""

    try:
        asin = soup.find("th", text="ASIN").find_next_sibling("td").text.strip()
    except:
        asin = ""

    try:
        product_description = soup.find("div", id="productDescription").text.strip()
    except:
        product_description = ""

    try:
        manufacturer = soup.find("a", {"id": "bylineInfo"}).text.strip()
    except:
        manufacturer = ""

    return [description, asin, product_description, manufacturer]


with open("products.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Product URL", "Product Name", "Product Price", "Product Rating", "Number of Reviews", "Description", "ASIN", "Product Description", "Manufacturer"])

    for page_num in range(1, 21):
        url = f"https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_{page_num}"
        response = requests.get(url)
        soup = BeautifulSoup(response.content, "html.parser")

        product_list = soup.find_all("div", class_="s-result-item")

        for product in product_list:
            try:
                product_url = "https://www.amazon.in" + product.find("a", class_="a-link-normal")["href"]
                if not product_url.startswith("https://www.amazon.in"):
                    raise ValueError("Invalid product URL")
            except:
                product_url = ""

            if product_url:
                product_info = get_product_info(product_url)

                try:
                    product_name = product.find("h2", class_="a-size-mini").text.strip()
                except:
                    product_name = ""

                try:
                    product_price = product.find("span", class_="a-price-whole").text.strip()
                except:
                    product_price = ""

                try:
                    product_rating = product.find("span", class_="a-icon-alt").text.strip().split()[0]
                except:
                    product_rating = ""

                try:
                    num_reviews = product.find("span", class_="a-size-base").text.strip().split()[0]
                except:
                    num_reviews = ""

                writer.writerow([product_url, product_name, product_price, product_rating, num_reviews] + product_info)

                print("Product URL:", product_url)
                print("Product Name:", product_name)
                print("Product Price:", product_price)
                print("Product Rating:", product_rating)
                print("Number of Reviews:", num_reviews)
                print("Description:", product_info[0])
                print()

            

print("Scraping completed and saved to products.csv file.")
```
